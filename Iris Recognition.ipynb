```python
from urllib.request import urlretrieve
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# fetch dataset and save it locally
dataset = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
urlretrieve(dataset)
```

```python
# assign column names
col = ["sepal_length", "sepal_width", "petal_length", "petal_width", "type"]
```

```python
# read dataset
iris = pd.read_csv(dataset, sep=',', names=col)
```

```python
# access dataframes by 'type 'column'
iris_setosa = iris.loc[iris['type'] == 'Iris-setosa']
iris_virginica = iris.loc[iris['type'] == 'Iris-virginica']
iris_versicolor = iris.loc[iris['type'] == 'Iris-versicolor']
```

```python
sns.FacetGrid(iris, hue='type', size=3).map(sns.distplot, 'petal_length').add_legend()
sns.FacetGrid(iris, hue='type', size=3).map(sns.distplot, 'petal_width').add_legend()
sns.FacetGrid(iris, hue='type', size=3).map(sns.distplot, 'sepal_length').add_legend()
sns.FacetGrid(iris, hue='type', size=3).map(sns.distplot, 'sepal_width').add_legend()
```

```python
# Probability density function & cumulative density function for iris setosa
counts, bin_edge = np.histogram(iris_setosa['petal_length'], bins=10, density=True)
pdf = counts/(sum(counts))
```

```python
cdf = np.cumsum(pdf)
plt.plot(bin_edge[1:], pdf)
plt.plot(bin_edge[1:], cdf)
plt.show()
```

```python
# iris virginica
counts, bin_edges = np.histogram(iris_virginica["petal_length"],bins=10, density=True)
pdf = counts/(sum(counts))
```

```python
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
```

```python
# iris versicolor
counts, bin_edges = np.histogram(iris_versicolor["petal_length"],bins=10, density=True)
pdf = counts/(sum(counts))
```

```python
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
```
